# DataFrame visualization

```python
url = "https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv"
```

```python
df = pd.read_csv(url)
df
```

```python
df = df.dropna(axis=1)
df
```

```python
len(df["terapia_intensiva"])
```

```python
import matplotlib.pyplot as plt
plt.plot(df["terapia_intensiva"])
```

```python
df.loc[:,"nuovi_in_terapia_intensiva"] = df["terapia_intensiva"].diff().fillna(0)
```

```python
plt.plot(df["nuovi_in_terapia_intensiva"])
```

```python
plt.plot(df["terapia_intensiva"],label="terapia_intensiva")
plt.plot(df["totale_positivi"], label="totale positivi")
plt.legend()
```

```python
plt.plot(df["terapia_intensiva"],label="terapia_intensiva")
plt.plot(df["totale_positivi"],label="totale positivi")
plt.legend()
plt.xlim(0,400)
plt.ylim(0,50000)
```

```python
percentage = df["terapia_intensiva"]/df["totale_positivi"]*100
percentage
```

```python
df.loc[:,"percentuale_terapia_intensiva"] = percentage
```

```python
plt.plot(df["percentuale_terapia_intensiva"],c="red")
```

```python
plt.hist(df["ricoverati_con_sintomi"])
plt.savefig("histogramma.png")
```

```python
def percentuale_positivi(df):
    
    df.loc[:,"nuovi_tamponi"] = df["tamponi"].diff().fillna(0)
    df.loc[:,"percentuale_positivi"] = df["nuovi_positivi"]/df["nuovi_tamponi"]*100
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df = df.fillna(0)
    
    return df
```

```python
df = percentuale_positivi(df)
df["percentuale_positivi"]
```
